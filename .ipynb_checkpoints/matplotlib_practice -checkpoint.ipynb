```python
#pie chart plt.pie()
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
explode_value = (0,0,0,0,0,0,0.2,0,0,0,0,0)
plt.pie(y_axis, labels=x_axis, explode=explode_value, autopct='%.1f%%')
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
#object oriented interfact to create pie chart 
fig, ax = plt.subplots(figsize=(8,8))
explode_value = (0,0,0.2,0,0,0,0.2,0,0,0,0,0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
ax.pie(y_axis, labels=x_axis, colors=colors, shadow=True, explode=explode_value, autopct='%.1f%%')
ax.invert_xaxis()
plt.show()

```

```python
#import statistics dependency and adding plt.errorbar() to calculate standard deviation 
%matplotlib 
import matplotlib.pyplot as plt 
import statistics 
```

```python
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
#calculate the standard deviation of y_axis 
stdev = statistics.stdev(y_axis)
```

```python
yerr = stdev 
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
fig,ax=plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
plt.bar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
fig,ax=plt.subplots()
ax.bar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
#using plt.xticks(np.arange(, step=5.0)) to create increments within a range (step is the increment value)
import numpy as np
plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0,51, step=5.0))
plt.gca().invert_yaxis()
plt.show()
```

```python
#adding ticks with object oriented interface 
fig,ax = plt.subplots()
ax.barh(x_axis, y_axis)
ax.xaxis.set_ticks(np.arange(0,51, step=5.0)) #or ax.set_xticks(np.arange(0,51, step=5.0))
```

```python
#add minor ticks between major ticks via x.xaxis.set_minor_locator()
from matplotlib.ticker import MultipleLocator 
#increase the size of the plot
fig, ax=plt.subplots(figsize=(8,8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0,51, step=5.0))
#create minor ticks on a increment of 1. 
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
