```python
import matplotlib.pyplot as plt 
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis)
plt.show
```

```python
#create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
#quick graph with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
plt.plot(x_axis, y_axis)
plt.show()
```

```python
#create the plot and add a label for the legend 
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
plt.subplots(figsize=(20, 10))
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.ylim(0, 45)
plt.xlim("Jan", "Jun")
plt.title("PyBer Fare by Month")
plt.grid()
plt.legend()
```

```python
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
ax.plot(x_axis, y_axis,  marker="^", color="green", linewidth=2, label='City of Boston')
ax.set_ylim(0, 45)
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.set_title("PyBer Fare by Month")
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#create the plot 
plt.bar(x_axis, y_axis)
```

```python
#annotation the bar chart (vertical)
plt.bar(x_axis, y_axis, color="green", label='Boston')
plt.xlabel('Date')
plt.ylabel('Fare($)')
plt.title('PyBer Fare by Month')
plt.legend()
```

```python
#horizontal bar chart 
plt.barh(x_axis, y_axis)
```

```python
#invert the yaxis to months in ascending order 
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

```python
plt.barh(x_axis, y_axis, color="magenta", label='Boston')
plt.xlabel('Date')
plt.ylabel('Fare($)')
plt.title('PyBer Fare by Month')
plt.gca().invert_yaxis()
plt.legend()
```

```python
# Set the x-axis to a list of strings for each month. (object_oriented method)
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#object_oriented method 
fig,ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color='cyan', label='Chicago')
ax.set_title('PyBer Fare by Month')
ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')
ax.invert_yaxis()
ax.legend()
```
